### Using Python, Pandas, Matplotlib and Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student_data = pd.read_csv("StudentsPerformance.csv")

student_data.head()
```

```python
student_data.info()
```

```python
student_data.isnull().sum()
```

```python
print((student_data[['math score', 'reading score', 'writing score']] < 0).sum())
```

```python
print("Average Math Score:", student_data['math score'].mean())

print("Average Reading Score:", student_data['reading score'].mean())

print("Average Writing Score:", student_data['writing score'].mean())
```

```python
student_data['total_score'] = (
    student_data['math score'] +
    student_data['reading score'] +
    student_data['writing score']
)

top_students = student_data.sort_values(
    by='total_score',
    ascending=False
).head(5)

top_students
```

```python
subject_averages = {
    'Math': student_data['math score'].mean(),
    'Reading': student_data['reading score'].mean(),
    'Writing': student_data['writing score'].mean()
}

plt.figure(figsize=(8,5))

plt.bar(
    subject_averages.keys(),
    subject_averages.values(),
    color=['skyblue', 'orange', 'green']
)

plt.title('Average Scores by Subject')

plt.xlabel('Subjects')
plt.ylabel('Average Score')

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.hist(
    student_data['math score'],
    bins=10
)

plt.title('Distribution of Math Scores')

plt.xlabel('Math Score')
plt.ylabel('Number of Students')

plt.show()
```

# Insights

- Reading and writing scores are higher on average compared to math scores.
- Top-performing students scored consistently across all subjects.
- The dataset contains valid academic scores without missing or negative values.
- Student performance distribution shows most students scored within the average range.
